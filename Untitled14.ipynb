```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, f_oneway
from docx import Document
from docx.shared import Inches
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from itertools import cycle

# Charger les données
data = pd.read_csv('C:/Users/HP/Documents/Test/Maternal_Health_Risk.csv')

# Créer un document Word pour le rapport
doc = Document()
doc.add_heading('Rapport d\'Analyse des Facteurs de Risque de la Mortalité Maternelle', level=1)

# 1. Visualisation des données
doc.add_heading('1. Visualisation des Données', level=2)
doc.add_paragraph("Aperçu des données :")
doc.add_paragraph(str(data.head()))
doc.add_paragraph("\nInformations sur les données :")
doc.add_paragraph(str(data.info()))
doc.add_paragraph("\nStatistiques descriptives :")
doc.add_paragraph(str(data.describe()))

# 2. Nettoyage des données
doc.add_heading('2. Nettoyage des Données', level=2)
doc.add_paragraph("Valeurs manquantes par colonne :")
doc.add_paragraph(str(data.isnull().sum()))
doc.add_paragraph("\nNombre de doublons : " + str(data.duplicated().sum()))

# 3. Statistique descriptive
doc.add_heading('3. Statistique Descriptive', level=2)
plt.figure(figsize=(10, 6))
sns.countplot(x='RiskLevel', data=data, palette='viridis', hue='RiskLevel', legend=False)
plt.title("Distribution des niveaux de risque")
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.savefig('risk_level_distribution.png')
doc.add_picture('risk_level_distribution.png', width=Inches(6))
doc.add_paragraph("La distribution des niveaux de risque montre que la majorité des femmes ont un niveau de risque modéré.")

# 4. Analyse univariée
doc.add_heading('4. Analyse Univariée', level=2)
for column in data.columns[:-1]:  # Exclure la variable cible
    plt.figure(figsize=(8, 4))
    sns.histplot(data[column], kde=True, color='blue')
    plt.title(f'Distribution de {column}')
    plt.savefig(f'{column}_distribution.png')
    doc.add_picture(f'{column}_distribution.png', width=Inches(6))
    doc.add_paragraph(f"La distribution de {column} montre que...")

# 5. Analyse bivariée
doc.add_heading('5. Analyse Bivariée', level=2)
for column in data.columns[:-1]:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='RiskLevel', y=column, data=data, palette='viridis', hue='RiskLevel', legend=False)
    plt.title(f'{column} par niveau de risque')
    plt.savefig(f'{column}_vs_risk_level.png')
    doc.add_picture(f'{column}_vs_risk_level.png', width=Inches(6))
    doc.add_paragraph(f"La relation entre {column} et le niveau de risque montre que...")

# Test statistique (ANOVA pour comparer les moyennes entre groupes)
doc.add_heading('6. Tests Statistiques', level=2)
for column in data.columns[:-1]:
    groups = [data[column][data['RiskLevel'] == level] for level in data['RiskLevel'].unique()]
    f_stat, p_value = f_oneway(*groups)
    doc.add_paragraph(f"ANOVA pour {column}: F-stat = {f_stat:.2f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        doc.add_paragraph(f"La variable {column} montre une différence significative entre les groupes de risque.")
    else:
        doc.add_paragraph(f"La variable {column} ne montre pas de différence significative entre les groupes de risque.")

# 7. Modélisation
doc.add_heading('7. Modélisation', level=2)

# Séparation des données
X = data.drop('RiskLevel', axis=1)
y = data['RiskLevel']  # On conserve les noms des modalités
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalisation des données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Fonction pour évaluer les modèles
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None

    # Matrice de confusion
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', 
                xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title(f'Matrice de confusion - {model_name}')
    plt.savefig(f'confusion_matrix_{model_name}.png')
    doc.add_picture(f'confusion_matrix_{model_name}.png', width=Inches(4))
    doc.add_paragraph(f"Matrice de confusion pour {model_name} :")

    # Rapport de classification
    doc.add_paragraph(f"Rapport de classification pour {model_name} :")
    report = classification_report(y_test, y_pred, target_names=model.classes_, output_dict=True)
    doc.add_paragraph(str(report))

    # Courbe ROC (si applicable et pour les problèmes multiclasses)
    if y_proba is not None and len(np.unique(y)) > 2:
        n_classes = len(np.unique(y))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(y_test == model.classes_[i], y_proba[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        plt.figure(figsize=(8, 6))
        colors = cycle(['blue', 'red', 'green', 'orange', 'purple'])
        for i, color in zip(range(n_classes), colors):
            plt.plot(fpr[i], tpr[i], color=color, lw=2,
                     label=f'Classe {model.classes_[i]} (AUC = {roc_auc[i]:.2f})')
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlabel('Taux de faux positifs')
        plt.ylabel('Taux de vrais positifs')
        plt.title(f'Courbe ROC multiclasse - {model_name}')
        plt.legend(loc="lower right")
        plt.savefig(f'roc_curve_{model_name}.png')
        doc.add_picture(f'roc_curve_{model_name}.png', width=Inches(6))
        doc.add_paragraph(f"Courbe ROC multiclasse pour {model_name} :")

    # Importance des caractéristiques (si applicable)
    if hasattr(model, "feature_importances_"):
        feature_importances = pd.Series(model.feature_importances_, index=X.columns)
        feature_importances.sort_values(ascending=False).plot(kind='bar', title=f'Importance des caractéristiques - {model_name}')
        plt.savefig(f'feature_importance_{model_name}.png')
        doc.add_picture(f'feature_importance_{model_name}.png', width=Inches(6))
        doc.add_paragraph(f"Importance des caractéristiques pour {model_name} :")

    return model

# Modèles à tester
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "SVM": OneVsRestClassifier(SVC(kernel='linear', probability=True, random_state=42))
}

# Évaluation des modèles
for name, model in models.items():
    doc.add_heading(f'7.1 {name}', level=3)
    trained_model = evaluate_model(model, X_train, X_test, y_train, y_test, name)

# Validation croisée pour le meilleur modèle
doc.add_heading('7.4 Validation Croisée', level=3)
best_model = RandomForestClassifier(random_state=42)
cv_scores = cross_val_score(best_model, X, y, cv=5, scoring='accuracy')
doc.add_paragraph(f"Scores de validation croisée (Random Forest) : {cv_scores}")
doc.add_paragraph(f"Moyenne des scores de validation croisée : {np.mean(cv_scores):.2f}")

# Sauvegarder le rapport Word
doc.save('Analyse_Mortalite_Maternelle.docx')
```
